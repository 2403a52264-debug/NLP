# We have took paragraph

```python
paragraph = "Natural language processing (NLP) is a branch of artificial intelligence that focuses on helping computers understand, interpret, and generate human language in a useful way. It combines concepts from linguistics and computer science so that machines can work with text and speech, like recognizing spoken commands, translating between languages, or answering questions. NLP is used in everyday tools such as virtual assistants, chatbots, spam filters, and language translators to make interactions with technology more natural and efficient."
```

# Here we are counting how many words were there in the paragraph

```python
count = 1
for char in paragraph:
  if char == " ":
    count += 1
print(count)
```

# Here we need to convert the whole paragraph into lowercase

```python
lower_paragraph = paragraph.lower()
print(lower_paragraph)
```
